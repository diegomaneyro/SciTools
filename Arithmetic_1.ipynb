```python
# Suma de matrices
a = np.array([[12, 23, 33], [40, 5, 6], [7, 8, 9]])
b = np.array([[9, 8, 7], [6, 52, 4], [3, 21, 10]])

c = a + b

print("La suma de la matriz a y b da la matriz c")

#Suma de matrices
print("Matriz: c\n",c)

# Transpuesta de la matriz c
print(f"La transpuesta de la matriz c es:\n{c.T}")

#deteminante de la matriz c
determinante = int(round(np.linalg.det(c)))
print(f"El determinante de la matriz c es: {determinante}")
```
