**Aritmetica**


```python
"Prueba de operaciones básicas aritméticas en Python"

a = 45
b = 21

print("A + B es: ", a + b)
print(f"A - B es: {a - b}")
print(f"A * B es: {a * b}")
print(f"A / B es: {a / b:#.2f}")
```

```python
print("=== Módulo ===")

respuesta = input("¿Desea probar el módulo? (s/n): ").lower()

while respuesta == "s":

    a = int(input("Ingresa un número: "))

    if a % 2 == 0:
        print(f"{a} es par")
    else:
        print(f"{a} es impar")

    respuesta = input("¿Desea probar otro número? (s/n): ").lower()

print("No se probará el módulo.")
```
